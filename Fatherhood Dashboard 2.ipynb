The initial cleaned dataset was merged with an additional dataset for more points of analysis to use in Tableau dashboards. 

The initial dashboard also revealed limitations in the original wide-format survey structure. 

The dataset was also reshaped into a long-format analytical model in Python to improve scalability, simplify pre/post comparisons, and support more flexible visualization design in Tableau/Power BI.

```python
import pandas as pd
import numpy as np
```

```python
fatherhood = pd.read_csv("Fatherhood Cleaned.csv")

map = pd.read_csv("Fatherhood Map Data.csv")
```

## Convert Client ID to string in both datasets to allow merge

```python
fatherhood["Client ID"] = (
    fatherhood["Client ID"]
    .astype("Int64")
    .astype("string")
)

map["Client ID"] = (
    map["Client ID"]
    .astype("Int64")
    .astype("string")
)
```

## Merge datasets

```python
fatherhood_merge = fatherhood.merge(
    map,
    on="Client ID",
    how="left",
    validate="one_to_one"
)
```

```python
print(fatherhood['Client ID'].dtype)
print(map['Client ID'].dtype)
```

```python
fatherhood_merge = fatherhood.merge(
    map,
    on="Client ID",
    how="left",
    validate="one_to_one"
)
```

```python
fatherhood_merge.to_csv('Fatherhood Cleaned v2.csv', index=False)
```
